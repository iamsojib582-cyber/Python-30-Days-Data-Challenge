```python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path(r"C:\Users\England\Documents\Downloads\Telco_Cusomer_Churn.csv")
OUTPUT_FILE = Path("Telco_Customer_Churn_cleaned.csv")


def clean_telco_churn(input_file: Path, output_file: Path) -> pd.DataFrame:
    df = pd.read_csv(input_file)

    df.columns = df.columns.str.strip()

    df = df.drop_duplicates(subset="customerID").copy()

    text_columns = df.select_dtypes(include=["object", "string"]).columns
    for column in text_columns:
        df[column] = df[column].astype(str).str.strip()

    df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
    df["TotalCharges"] = df["TotalCharges"].fillna(0)

    df["SeniorCitizen"] = pd.to_numeric(df["SeniorCitizen"], errors="coerce").fillna(0).astype(int)
    df["tenure"] = pd.to_numeric(df["tenure"], errors="coerce").fillna(0).astype(int)
    df["MonthlyCharges"] = pd.to_numeric(df["MonthlyCharges"], errors="coerce")

    yes_no_columns = [
        "Partner",
        "Dependents",
        "PhoneService",
        "PaperlessBilling",
        "Churn",
    ]
    for column in yes_no_columns:
        if column in df.columns:
            df[column] = df[column].replace({"yes": "Yes", "no": "No"})

    df.to_csv(output_file, index=False)
    return df


if __name__ == "__main__":
    cleaned_df = clean_telco_churn(INPUT_FILE, OUTPUT_FILE)

    print(f"Cleaned rows: {len(cleaned_df)}")
    print(f"Columns: {len(cleaned_df.columns)}")
    print(f"Blank TotalCharges after cleaning: {cleaned_df['TotalCharges'].isna().sum()}")
    print(f"Output saved to: {OUTPUT_FILE.resolve()}")
```

```python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path(r"C:\Users\England\Documents\Downloads\Telco_Cusomer_Churn.csv")
OUTPUT_FILE = Path("Telco_Customer_Churn_cleaned_yes_no.csv")

df = pd.read_csv(INPUT_FILE)

df.columns = df.columns.str.strip()

df = df.drop_duplicates(subset="customerID").copy()

text_columns = df.select_dtypes(include=["object", "string"]).columns
for column in text_columns:
    df[column] = df[column].astype(str).str.strip()

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"] = df["TotalCharges"].fillna(0)

df["SeniorCitizen"] = pd.to_numeric(df["SeniorCitizen"], errors="coerce").fillna(0).astype(int)
df["tenure"] = pd.to_numeric(df["tenure"], errors="coerce").fillna(0).astype(int)
df["MonthlyCharges"] = pd.to_numeric(df["MonthlyCharges"], errors="coerce")

service_columns = [
    "MultipleLines",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
]

for column in service_columns:
    df[column] = df[column].replace({
        "No internet service": "No",
        "No phone service": "No"
    })

yes_no_columns = [
    "Partner",
    "Dependents",
    "PhoneService",
    "PaperlessBilling",
    "Churn",
] + service_columns

for column in yes_no_columns:
    df[column] = df[column].replace({
        "yes": "Yes",
        "no": "No"
    })

df.to_csv(OUTPUT_FILE, index=False)

print("Cleaning complete.")
print(f"Saved to: {OUTPUT_FILE.resolve()}")
```
