```python
high_sales = []
with open("sales_data.csv", "r") as file:
    lines = file.readlines()
    header=lines[0]
    high_sales.append(header)
    for line in lines[1:]:
        values=line.strip()
        try:
            purchase = float(values[4]) 
            if purchase >= 5000:
                high_sales.append(values)
        except ValueError:
            pass
print(f"High sales customers: {len(high_sales) - 1}")
for row in high_sales:
    print(row)
```

```python
city_sales = {}

with open("sales_data.csv", "r") as file:
    lines = file.readlines()
    for line in lines[1:]:
        values = line.strip().split(",")
        city = values[3]
        try:
            purchase = float(values[4])
            if city not in city_sales:
                city_sales[city] = 0
            city_sales[city] += purchase
        except ValueError:
            pass

print("Sales by City:")
for city, total in city_sales.items():
    print(f"{city.capitalize()}: Total Sales = ${total:.2f}")
```

```python
customer_purchases = {}

with open("sales_data.csv", "r") as file:
    lines = file.readlines()
    for line in lines[1:]:
        values = line.strip().split(",")
        customer_id = values[0]
        if customer_id not in customer_purchases:
            customer_purchases[customer_id]=0
            customer_purchases[customer_id]=1
print("Customer with Multiple Purchases:")
for customer_id, count in customer_purchases.items ():
    if count >1:
        print(f" {customer_id}: {count} purchases")
        
            
```

```python
def find_customer_by_email(email_to_find):
    with open("sales_data.csv", "r") as file:
        lines = file.readlines()
        header=lines[0].strip().split(",")
        for line in lines[1:]:
            values=line.strip().split()
            row_dict=dict(zip(header, values))
            if row_dict.get ("Email")==email_to_find:
                return row_dict
            return None
customer = find_customer_by_email("")
if customer:
    print(customer)
else:
    print("Customer Not Found")
      
```

```python
with open("sales_data.csv", "r") as file:
    lines = file.readlines()
    header = lines[0].strip().split(",")
print("SALES DATA SUMMARY REPORT")
print("=" * 50)
print(f"Columns: {', '.join(header)}")
print(f"Total Rows: {len(lines) - 1}")
total_sales = 0
count = 0
cities = set()
for line in lines[1:]:
    values = line.strip().split(",")
    try:
        purchase = float(values[4])
        total_sales += purchase
        count += 1
        city = values[3]
        cities.add(city)
    except ValueError:
        pass
average=total_sales / count if count > 0 else 0
print(f"\nMetrics:")
print(f"  Total Sales: ${total_sales:.2f}")
print(f"  Average Sale: ${average:.2f}")
print(f"  Transactions: {count}")
print(f"  Cities: {len(cities)}")
print(f"  City List: {', '.join(sorted(cities))}")
print("=" * 50)
```

```python
from datetime import datetime
monthly_customers={}
with open("sales_data.csv", "r") as file:
    lines = file.readlines()
    for line in lines[1:]:
        values = line.strip().split(",")
        if len(values) >= 6:
            date_str = values[5]
            try:
                date_obj = datetime.strptime(date_str, "%Y-%m-%d")
                month_key = date_obj.strftime("%Y-%m") 
                if month_key not in monthly_customers:
                    monthly_customers[month_key] = 0
                    monthly_customers[month_key] += 1
            except ValueError:
                pass
print("Customers by Month:")
for month in sorted(monthly_customers.keys()):
    count = monthly_customers[month]
    print(f"  {month}: {count} customers")
                    
```

```python
valid_count = 0
error_count = 0
errors = []
with open("sales_data.csv", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines[1:], start=2):
    values = line.strip().split(",")
    if len(values) < 6:
        error_count += 1
        errors.append(f"Line {i}: Missing columns")
        continue
    if "@" not in values[2]:
        error_count += 1
        errors.append(f"Line {i}: Invalid email")
        continue
    try:
        float(values[4])
    except ValueError:
        error_count += 1
        errors.append(f"Line {i}: Invalid purchase amount")
        continue
    valid_count += 1
    print(f"Valid rows: {valid_count}")
print(f"Error rows: {error_count}")
print("\nErrors:")
for error in errors[:10]:  # Show first 10
    print(f"  {error}")
     
```
