```python
with open ("C:\\Users\\England\\Documents\\Downloads\\messy_customers_data.csv", "r") as file:
    lines=file.readlines ()
    header=lines[0].strip().split(",")
    for i, line in enumerate(lines[1:],start=2):
        values=line.strip().split(",")
        if len(values)< len(header):
            print(f"Line{i}:Missing Columns")
        for j, val in enumerate(values):
            if not val.strip():
                print(f"Line {i}, Column {header[j]}: Empty value")
```

```python
with open ("C:\\Users\\England\\Documents\\Downloads\\messy_customers_data.csv", "r") as file:
     lines=file.readlines ()
     header=lines[0]
     seen=set()
     unique=[header]
for line in lines [1:]:
    customer_id=line.split(",") [0]
    if customer_id not in seen:
        seen.add(customer_id)
        unique.append(line)
with open ("cleaned_customers.csv", "w") as file:
    file.writelines(unique)
print(f"Removed duplicates! Saved {len(unique)-1} unique records.")  
```

```python
with open ("C:\\Users\\England\\Documents\\Downloads\\messy_customers_data.csv", "r") as file:
    lines=file.readlines ()
    total_sales=0
    valid_count=0
    for line in  lines[1:]:
        values=line.strip().split(",")
        if len(values) >=5 :
            try:
                amount=float(values[4])
                if amount > 0:
                    total_sales+= amount
                    valid_count+=1
            except ValueError:
                print(f"Invalid Amount:{values[4]}")
print(f"Total valid sales: ${total_sales}")
print(f"Valid records: {valid_count}")
```

```python
with open ("C:\\Users\\England\\Documents\\Downloads\\messy_customers_data.csv", "r") as file:
    lines=file.readlines ()
    header=lines[0]
    high_value=[header]
    for line in lines [1:]:
        values=line.strip().split(",")
        if len(values)>5:
            try:
                amount=float(values[4])
                if amount >=5000:
                    high_value.append(line)
            except ValueError:
                pass
with open ("high_value_customers.csv", "w") as file:
    file.writelines(high_value)
print(f"Found {len(high_value)-1} high-value customers!")
            
```

```python
# Email Validation
with open ("C:\\Users\\England\\Documents\\Downloads\\messy_customers_data.csv", "r") as file:
    lines=file.readlines ()
errors=[]
for i, line in enumerate (lines[1:], start=2):
    values=line.strip().split(",")
    if len(values)>=3:
        email=values[2]
        if email and ("@" not in email or "." not in email):
            errors.append(f"Line {i}:Invalid Email :{email}")
print(f"Found {len(errors)}email errors")
for error in errors:
    print(f"{  error}")
               

    
```
