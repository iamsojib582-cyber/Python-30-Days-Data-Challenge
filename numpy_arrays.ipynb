```python
import numpy as np
arr = np.array([1, 2, np.nan, 4, 5, np.nan, 7])
print(arr)
print(np.isnan(arr))
```

```python
arr_clean = arr[~np.isnan(arr)]
print(arr_clean)
```

```python
nan_count = np.sum(np.isnan(arr))
print(f"Missing values: {nan_count}")
```

```python
arr_filled = np.where(np.isnan(arr), 0, arr)
print(arr_filled)
```

```python
mean_value = np.nanmean(arr)
arr_filled = np.where(np.isnan(arr), mean_value, arr)
print(arr_filled)
```

```python
arr = np.array([10, 12, 15, 18, 20, 100, 22, 25])
mean = np.mean(arr)
std = np.std(arr)
outliers = np.abs(arr - mean) > 2 * std
arr_clean = arr[~outliers]
print(f"Original: {arr}")
print(f"Cleaned: {arr_clean}")
```

```python
arr = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5])
unique_arr = np.unique(arr)
print(unique_arr)
unique, counts = np.unique(arr, return_counts=True)
for val, count in zip(unique, counts):
    print(f"{val}: appears {count} times")
```

```python
arr = np.array([10, 20, 30, 40, 50])
arr_min = np.min(arr)
arr_max = np.max(arr)
arr_normalized = (arr - arr_min) / (arr_max - arr_min)
print(arr_normalized)
```

```python
arr = np.array([10, 20, 30, 40, 50])
mean = np.mean(arr)
std = np.std(arr)
arr_standardized = (arr - mean) / std
print(arr_standardized)
```

```python
import numpy as np
sales = np.array([100, 150, np.nan, 200, 180, 5000, 220, 190, 250, np.nan])
print(f"Original data: {sales}")
nan_mask = np.isnan(sales)
print(f"NaN values: {np.sum(nan_mask)}")
sales_no_nan = sales[~nan_mask]
mean = np.mean(sales_no_nan)
std = np.std(sales_no_nan)
outlier_mask = np.abs(sales_no_nan - mean) > 3 * std
sales_clean = sales_no_nan[~outlier_mask]
print(f"Cleaned data: {sales_clean}")
print(f"Mean: {np.mean(sales_clean):.2f}")
print(f"Std Dev: {np.std(sales_clean):.2f}")
```

```python
import numpy as np
temps = np.array([20, 22, 25, -999, 23, 24, 1000, 22, 21, 20])
temps_clean = np.where((temps < 0) | (temps > 50), np.nan, temps)
temps_valid = temps_clean[~np.isnan(temps_clean)]
print(f"Cleaned temperatures: {temps_valid}")
print(f"Average: {np.mean(temps_valid):.2f}°C")
```

```python
ratings = np.array([4.5, 3.2, 4.8, 5.1, 2.9, 4.2, 6.0, 3.8, 4.1])
ratings_clean = np.where((ratings >= 1) & (ratings <= 5), ratings, np.nan)
ratings_valid = ratings_clean[~np.isnan(ratings_clean)]
print(f"Valid ratings: {ratings_valid}")
print(f"Average rating: {np.mean(ratings_valid):.2f}")
```

```python
import numpy as np
arr = np.array([10, 20, 15, 25, 30, 22, 18])
print(f"Count: {len(arr)}")
print(f"Sum: {np.sum(arr)}")
print(f"Mean: {np.mean(arr):.2f}")
print(f"Median: {np.median(arr):.2f}")
print(f"Mode: {np.bincount(arr).argmax()}")
print(f"Std Dev: {np.std(arr):.2f}")
print(f"Variance: {np.var(arr):.2f}")
print(f"Min: {np.min(arr)}")
print(f"Max: {np.max(arr)}")
print(f"Range: {np.max(arr) - np.min(arr)}")
print(f"Quartiles: {np.percentile(arr, [25, 50, 75])}")
```

```python
sales = np.array([100, 150, 200, 250, 300])
profit = np.array([20, 35, 50, 65, 80])
correlation = np.corrcoef(sales, profit)[0, 1]
print(f"Correlation: {correlation:.3f}")
covariance = np.cov(sales, profit)[0, 1]
print(f"Covariance: {covariance:.2f}")
```
